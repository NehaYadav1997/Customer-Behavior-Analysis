```python
#intead of filling nulls with overall mean, null is filled by each category's mean
# this will help in maintining the categorical rating
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

```python
# creating a column age_group to group customers into 4 age groups which will help in understanding patterns accros the various age group  
```

```python
labels = ['Yound Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(15)
```

```python
# creating feature purchase_frequency_days by converting text into numbers to make analysis easier
```

```python
frequency_mapping ={
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
